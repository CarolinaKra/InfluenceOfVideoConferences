# Influence of Video Conferencing on Society during the Pandemic

# Research topic and background

## Introduction

Since the beginning of the Covid-19 pandemic, strict lockdowns have been imposed around the world. Schools and business closed, people were not allowed to gather for non-essential purposes such as and family and social meetings, cultural events, or sports activities. Some of the normal face-to-face meetings were replaced by video conferences on different software such as Zoom. The article dated 10/05/2020 from The Guardian, explains that “video conferencing apps [were] downloaded 62 million times over 10 days in March”. They said that business and organisations were able to continue what they are doing but remotely thanks to this software. Despite the benefits of video conferencing, they stressed that the physical companionship and the body language are not possible to replicate with online communication.
The research objective is to investigate the influence of video conferencing (VC) on society during the pandemic and the motivation for this research is to understand the future consequences of using VC as a replacement for face-to-face meetings.

### Research Questions:
1.	Explore the use of video conference meetings (VCM) by the different population groups as a replacement of face-to-face meetings during the pandemic.
2.	Understand people’s satisfaction with VC in comparison to face-to-face meetings. 
3.	Evaluate the personal benefits of VC during the pandemic.
4.	Attempt to predict the use of VC in the future post-pandemic.

The first two research questions help understand the overall present use of VC and the people’s general feeling about using the VC.
The third research questions will bring a more depth inside about the personal benefit from VC and the fourth will help predict the future use. These topics have not been analysed yet and could bring benefit to society and could serve as motivation for new projects or services. 

### Population:
The survey was designed for the general population, for everyone over 13.
It was distributed to friends, family, colleagues, fellow students, and others on social media. Some participants shared it with others. Even though I had a large distribution, the population was biased towards people on similar age range as myself.

### Domain Concepts
Video Conference meeting: It is a live, visual, and audial connection between two or more remote parties over the internet. For this research, we consider it as any type of digital meeting where participants can listen to each other and can see each other and/or share contents on a screen.


# Survey Design
It was a self-administered online survey with 3 pages:
1.	The Introduction. Research objective and researcher presentation. Domain concept and instructions were provided
2.	The survey itself. First, the demographic questions were asked, then questions about the use of VC and general satisfaction with VC, then the personal benefit questions. Lastly, the questions related to the future.
3.	The Third page gave acknowledgements for participating in the survey. 

### Operational Definitions
The survey is composed mainly of multiple answer questions where, on some of them, it is expected from the participant to select one answer, and some of them, they can select multiple answers. There is a (yes/no) question, which is followed by another question in the case they answer yes. On some of the questions, people can select “other” answer and specify it.

```python
#To extract the advantages from one single column with long text into simpler advantages, it was more complex comparing to the
#meeting purpose variable, because there was a coma in the text. Hence I needed to find key words from each category and create
#a separate dataframe with one-hot encoding regarding the number of advantages 
advantages=np.zeros((df.shape[0],5))
for (i,val) in enumerate(df['Advantages']):
    for word in val.split():
        if (re.findall('Meeting',word)):
            advantages[i,0]=1
        if (re.findall('digital',word)):
            advantages[i,1]=1
        if (re.findall('information',word)):
            advantages[i,2]=1
        if (re.findall('money',word)):
            advantages[i,3]=1
        if (re.findall('Other',word)):
            advantages[i,4]=1
adv_cols=['meeting with people in different places','digital tools','sharing information','less expenditure','Adv_other']
AdvanDF=pd.DataFrame(advantages, columns=adv_cols)
AdvanDF.index.name='id'
AdvanDF.head()
```

```python
# I created another variable that sums the number of selected advantages
number_of_advantages=AdvanDF.sum(axis=1)
number_of_advantages.head()
```

```python
#For the disadvantages variable, I used the same technique as for the advantages 
disadvantages=np.zeros((df.shape[0],5))
for (i,val) in enumerate(df['Disadvantages']):
    for word in val.split():
        if (re.findall('Decrease',word)):
            disadvantages[i,0]=1
        if (re.findall('Deterioration',word)):
            disadvantages[i,1]=1
        if (re.findall('distraction',word)):
            disadvantages[i,2]=1
        if (re.findall('efficiency',word)):
            disadvantages[i,3]=1
        if (re.findall('Other',word)):
            disadvantages[i,4]=1
dis_cols=['Decrease in personal and affective contact','decrease in non-verbal communication','increase in distraction',
          'lower efficiency','Dis_other']
DisadvanDF=pd.DataFrame(disadvantages, columns=dis_cols)
DisadvanDF.index.name='id'
DisadvanDF.head()
```

```python
#I created as well another variable with the number of disadvantages
number_of_disadvantages=DisadvanDF.sum(axis=1)
number_of_disadvantages.head()
```

```python
#For the type of personal development, I used the same pre-processing as for the meeting purpose variable 
type_of_development=df['Type_of_Development'].str.split(',', expand=True)
type_of_development.head()
# This question was only answer by the people who selected yes on stimulation for personal development hence many people
#haven't answer 
```

Since the beginning of the pandemic, VCM has been part of the daily activities in society. It helps people stay connected and continue their occupation such as studying and working as well as having leisure activities.

We could see that most of the people use VC every day, meet between 4 and 7 people in a meeting which can take between 30 minutes and one hour. We learnt how the youngsters and the elders use VC differently than the intermediate age groups.

Even though people are disappointed by the decrease on personal and affective contact, they are somewhat satisfied with VC and find it useful that they can meet with many people that are in different places and they don't need to spend money and time travelling.

We have seen that VC kept the society connected which helped them cope with social distancing by a reasonable amount. In addition, for almost a third of the population, VC has stimulated their personal development, mostly by developing their technological skills.

We can predict that in the future post-pandemic, the population will use VC less frequent than nowadays, and it will continue to be an important tool for work and study purposes.

We can conclude that VC brings a great benefit for the society in the way that people can connect and the way that individuals can develop, especially during the pandemic.

This work could inspire the development of novel VC applications which could help connect between people around the globe and help more people develop personally and professionally.
